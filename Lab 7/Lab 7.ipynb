### Serving an MLflow Model from Model Registry

**Run this from command line**
`set MLFLOW_TRACKING_URI=http://localhost:5000` 
#use export MLFLOW_TRACKING_URI=http://localhost:5000 if in linux

## **Now run this command from command line**

make sure to write the different port number - other than the one you used while starting mlflow server/ui
This will do the model serving part

`mlflow models serve --model-uri models:/name_of_the_model/Production -p 1234 --no-conda`

```python
import requests
lst = X_test.values.tolist()
inference_request = {
        "dataframe_records": lst
}
endpoint = "http://localhost:1234/invocations"
response = requests.post(endpoint, json=inference_request)
print(response)
```
